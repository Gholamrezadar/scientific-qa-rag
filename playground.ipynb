```python
import requests

def fetch_wikipedia_summary(keyword, lang='en'):
    url = f"https://{lang}.wikipedia.org/api/rest_v1/page/summary/{keyword}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        print("Title:", data.get("title"))
        print("Summary:", data.get("extract"))
        print("URL:", data.get("content_urls", {}).get("desktop", {}).get("page"))
    else:
        print("Failed to fetch data:", response.status_code)

def fetch_wikipedia_content(keyword, lang='en'):
    url = f"https://{lang}.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "explaintext": True,
        "titles": keyword
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        pages = response.json().get("query", {}).get("pages", {})
        for page_id, page in pages.items():
            print("Title:", page.get("title"))
            print("Content:", page.get("extract"))
    else:
        print("Failed to fetch data:", response.status_code)
```

```python
fetch_wikipedia_summary("chemical potential")
```

```python
fetch_wikipedia_content("Quark")
```

```python
from ollama import chat
from ollama import ChatResponse
from src.prompts import get_answer_prompt 
import random
import time

model_name = 'gemma3:1b'

context = 'Gholamreza is a software engineer who likes machine learning.'
question = 'What is Gholamreza\'s job?'
choices = ['Software Engineer', 'Chef', 'Teacher', 'Doctor', 'Lawyer']
random.shuffle(choices)


prompt = get_answer_prompt(question, choices, context)
print("Prompt:")
print(prompt)
print('- '*40)
start_time = time.perf_counter()
response: ChatResponse = chat(model=model_name, messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
end_time = time.perf_counter()
print(f'Time taken: {end_time - start_time:.3f} seconds')
print("Response: ")
print(response['message']['content'])
print(f'Correct Answer: {chr(65+choices.index("Software Engineer"))}')
```

```python
from langchain_ollama import ChatOllama
from langchain.schema import HumanMessage
```

```python
from src.prompts import get_answer_prompt

model_name = "gemma3:1b"
chat_model = ChatOllama(model=model_name)

context = "Gholamreza is a software engineer who likes machine learning."
question = "What is Gholamreza's job?"
choices = ['Software Engineer', 'Chef', 'Teacher', 'Doctor', 'Lawyer']
random.shuffle(choices)

prompt = get_answer_prompt(question, choices, context)
print("Prompt:")
print(prompt)
print("- " * 40)

start_time = time.perf_counter()
response = chat_model([HumanMessage(content=prompt)])
end_time = time.perf_counter()

print(f"Time taken: {end_time - start_time:.3f} seconds")
print("Response:")
print(response.content)
print(f"Correct Answer: {chr(65 + choices.index('Software Engineer'))}")
```
